```python
import numpy as np
import pandas as pd
import scipy as sp
from scipy import stats

from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
```

### 3.7 추정

```python
fish = pd.Series([4.352982, 3.735304, 5.944617, 3.798326, 4.087688,
                  5.265985, 3.272614, 3.526691, 4.150083, 3.736104])
fish
```

- point etimation 점추정

```python
mu = np.mean(fish)
mu
```

```python
np.var(fish, ddof=1)
```

- Interval Estimation 구간추정\
추정값이 폭을 가지게 하는 추정법. 폭 계산에는 확률 개념 사용\
**신뢰계수** 란 구간추정의 폭에 대한 신뢰 정도를 확률로 표현한 것\
**신뢰구간** 이란 특정 신뢰계수를 만족하는 구간

- confidence interval 신뢰구간 

$ \frac{표본평균-모평균}{표준오차} $ \ $ 표준오차 = \frac{\sigma}{\sqrt{N}} $

```python
# 표준오차
se = np.std(fish, ddof=1) / np.sqrt(len(fish))
se
```

```python
# 신뢰구간 - 인수(alpha = 신뢰계수, df = 자유도, loc = 표본평균, scale = 표준오차)
stats.t.interval(alpha=0.95, df=len(fish)-1, loc=mu, scale=se)
```
